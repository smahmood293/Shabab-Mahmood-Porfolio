```python
file_path = '/kaggle/input/top-10000-spotify-songs-1960-now/top_10000_1960-now.csv'
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
print(df.columns)
```

```python
column_drop = ['Track URI', 'Artist URI(s)','Time Signature', 'Album URI', 'Album Artist URI(s)', 'Album Image URL', 'Disc Number', 'Track Number', 'Track Preview URL', 'ISRC', 'Added By', 'Added At', 'Album Genres', 'Label', 'Copyrights']
df = df.drop(columns = column_drop)
df.columns
```

```python
import math
df['Track Duration (min)'] = df['Track Duration (ms)']/60000
df['Track Duration (min)'] = df['Track Duration (min)'].apply(lambda x: math.ceil(x))
df = df.drop('Track Duration (ms)', axis = 1)
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
print(df.dtypes)
print(df.shape[0])
```

```python
df = df.drop_duplicates()
df.duplicated().sum()
```

```python
column_fill = ['Danceability', 'Energy', 'Key', 'Loudness', 'Mode', 'Speechiness', 'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo', 'Track Duration (min)']  
df[column_fill] = df[column_fill].apply(lambda col: col.fillna(col.mean()))
df['Album Release Date'] = df['Album Release Date'].fillna(df['Album Release Date'].mode()[0])
drop_columns = ['Track Name','Artist Name(s)', 'Album Name', 'Album Artist Name(s)', 'Artist Genres']
df = df.drop(drop_columns, axis = 1)
print(df.isnull().sum())
print(df.columns)
```

```python
df['Album Release Date'] = pd.to_datetime(df['Album Release Date'], errors='coerce')
df['Album Release Year'] = df['Album Release Date'].dt.year
df = df.drop('Album Release Date', axis = 1)
df = df.dropna(subset=['Album Release Year'])
print(df.isnull().sum())
print(df.shape[0])
```

```python
avg_metrics = df.groupby('Album Release Year').agg({
    'Danceability': 'mean',
    'Acousticness': 'mean',
    'Instrumentalness': 'mean',
    'Liveness': 'mean',
    'Speechiness': 'mean',
    'Energy': 'mean'}).reset_index()
plt.plot(avg_metrics['Album Release Year'], avg_metrics['Danceability'], label='Average Danceability')
plt.plot(avg_metrics['Album Release Year'], avg_metrics['Acousticness'], label='Average Acousticness')
plt.plot(avg_metrics['Album Release Year'], avg_metrics['Instrumentalness'], label='Average Instrumentalness')
plt.plot(avg_metrics['Album Release Year'], avg_metrics['Liveness'], label='Average Liveness')
plt.plot(avg_metrics['Album Release Year'], avg_metrics['Speechiness'], label='Average Speechiness')
plt.plot(avg_metrics['Album Release Year'], avg_metrics['Energy'], label='Average Energy')

# Add labels and legend
plt.xlabel('Year')
plt.ylabel('Average Value')
plt.title('Average Metrics per Year')
plt.legend()

plt.show()
```

```python
avg_track_time = df.groupby('Album Release Year')['Track Duration (min)'].mean().reset_index()

# Plot the data
plt.plot(avg_track_time['Album Release Year'], avg_track_time['Track Duration (min)'])
plt.xlabel('Year')
plt.ylabel('Average Track Duration (Minutes)')
plt.title('Average Track Duration per Year')
plt.show()
```

```python
avg_loudness = df.groupby('Album Release Year')['Loudness'].mean().reset_index()

# Plot the data
plt.plot(avg_loudness['Album Release Year'], avg_loudness['Loudness'])
plt.xlabel('Year')
plt.ylabel('Average Loudness')
plt.title('Average Loudness Over Time')
plt.show()
```

```python
avg_tempo = df.groupby('Album Release Year')['Tempo'].mean().reset_index()

# Plot the data
plt.plot(avg_tempo['Album Release Year'], avg_tempo['Tempo'])
plt.xlabel('Year')
plt.ylabel('Average Tempo')
plt.title('Average Tempo Over Time')
plt.show()
```
