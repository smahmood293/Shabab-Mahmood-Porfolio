```python
ten_best_player_OAT = pd.read_sql("""
                            SELECT DISTINCT Player.player_name, Player_Attributes.overall_rating
                            FROM Player
                            JOIN Player_Attributes ON Player.player_api_id = Player_Attributes.player_api_id
                            ORDER BY overall_rating DESC
                            LIMIT 10;
                            """, conn)
ten_best_player_OAT
```

```python
Highest_Potential_Youngsters = pd.read_sql("""
SELECT DISTINCT Player.player_name, Player.birthday, Player_Attributes.potential, Player_Attributes.overall_rating
FROM Player
JOIN Player_Attributes ON Player.player_api_id = Player_Attributes.player_api_id
WHERE strftime('%Y', 'now') - strftime('%Y', Player.birthday) <= 26
ORDER BY Player_Attributes.potential DESC
LIMIT 10;
""",conn)
Highest_Potential_Youngsters
```

```python
leagues_by_season = pd.read_sql("""SELECT Country.name AS country_name, 
                                        League.name AS league_name, 
                                        season,
                                        count(distinct stage) AS number_of_stages,
                                        count(distinct HT.team_long_name) AS number_of_teams,
                                        avg(home_team_goal) AS avg_home_team_scors, 
                                        avg(away_team_goal) AS avg_away_team_goals, 
                                        avg(home_team_goal-away_team_goal) AS avg_goal_dif, 
                                        avg(home_team_goal+away_team_goal) AS avg_goals, 
                                        sum(home_team_goal+away_team_goal) AS total_goals                                       
                                FROM Match
                                JOIN Country on Country.id = Match.country_id
                                JOIN League on League.id = Match.league_id
                                LEFT JOIN Team AS HT on HT.team_api_id = Match.home_team_api_id
                                LEFT JOIN Team AS AT on AT.team_api_id = Match.away_team_api_id
                                WHERE country_name in ('Spain', 'Germany', 'France', 'Italy', 'England')
                                GROUP BY Country.name, League.name, season
                                HAVING count(distinct stage) > 10
                                ORDER BY Country.name, League.name, season DESC
                                ;""", conn)
leagues_by_season
```

```python
tallest = pd.read_sql("""
                            SELECT DISTINCT(Player.player_name), Player.height, Player.weight
                            FROM Player
                            ORDER BY height DESC
                            LIMIT 10;
                            """, conn)
tallest
```

```python
heaviest = pd.read_sql("""
                            SELECT DISTINCT(Player.player_name), Player.height, Player.weight
                            FROM Player
                            ORDER BY weight DESC
                            LIMIT 10;
                            """, conn)
heaviest
```

```python
player_stats = pd.read_sql("""SELECT
    Player.player_name,
    Player.height,
    Player.weight,
    AVG(Player_Attributes.overall_rating) AS avg_overall_rating
FROM Player
JOIN Player_Attributes ON Player.id = Player_Attributes.id
GROUP BY Player.player_name, Player.height, Player.weight
ORDER BY avg_overall_rating DESC;""",conn)
player_stats
```
